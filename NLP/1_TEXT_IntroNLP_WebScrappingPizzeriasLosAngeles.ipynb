### Libraries

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
import nltk
```

### Yellow pages scrapping

```python
webpage = 'https://www.yellowpages.com/search?search_terms=pizza&geo_location_terms=Los+Angeles%2C+CA' 
    #simplemente es la consulta que hago en la web
```

```python
page = requests.get(webpage)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
test = soup.findAll('div', {'class':"info"})
```

```python
type(test)
```

```python
type(str(test))
```

### Get phone numbers

```python
list_url =nltk.word_tokenize(str(test))  ##TOKENIZAMOS TODO O QUE HA DEVUELTO DE LA CAPA DIV
```

```python
r_phone = re.compile("\d{3}-\d{4}") #coge una regla de cuando encuentre 3 numeros "-"" 4 numeros me lo recupere
    #podriamos filtrar por el div que se llama Phone
```

```python
### Phone Numbers
phone_numbers = list(set(r_phone.findall(str(test))))
```
