```python
# -*- coding: utf-8 -*-
"""
Created on Mon Mar 19 10:53:22 2018

@author: cristina.puente
"""
from bs4 import BeautifulSoup
import requests

page = requests.get("http://dataquestio.github.io/web-scraping-pages/simple.html")
print(page)
#print(page.content)
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
print(list(soup.children))
[print(type(item)) for item in list(soup.children)]
html = list(soup.children)[2]
print(list(html.children))

body = list(html.children)[3]
print(list(body.children))
p = list(body.children)[1]
print(p.get_text())

"""
Por etiquetas class e id

"""
print("USANDO LAS ETIQUETAS")
soup.find_all('p')
print(soup.find_all('p')[0].get_text())

"""
Web con capas

"""
page = requests.get("http://dataquestio.github.io/web-scraping-pages/ids_and_classes.html")
soup = BeautifulSoup(page.content, 'html.parser')
soup
texto=soup.find_all('p')
print("USANDO CAPAS")
for texto in texto:
    print(texto.get_text()) #LISTAR TODAS
print("USANDO ALGUNAS CAPAS")    
c=soup.find_all('p', class_='outer-text')
for c in c:
    print(c.get_text())
    
ident=soup.find_all(id="first")    
print("USANDO ID's") 
for ident in ident:
    print(ident.get_text())
    
print("WEATHER TABLE")
page = requests.get("https://www.timeanddate.com/weather/benin/cotonou")
soup = BeautifulSoup(page.content, 'html.parser')
            
data = []
table = soup.find('table', attrs={'id':'wt-48'})
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    print(cols)
    data.append([ele for ele in cols if ele])

```
